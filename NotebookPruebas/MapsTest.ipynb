```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
import gmaps
import gmaps.datasets
gmaps.configure(api_key='YOURAPIKEY') # Fill in with your API key
```

```python
blackshapes = pd.read_csv('blackshapes2014.v2.csv')
blackshapes.head(3)
```

```python
blackshapes_points = blackshapes[['lat', 'long', 'Address','numAccident']]
```

```python
myMap = folium.Map(location=(37.9943708, -1.1275828), zoom_start=4)
marker_cluster = MarkerCluster().add_to(myMap)
for index, point  in blackshapes_points.iterrows():
    folium.Marker(location = [point.lat, point.long],popup='Carretera : ' + str(point.Address) + '\n' + 'Accidentes: ' + str(int(point.numAccident)) ).add_to(marker_cluster)

display(myMap)
```

```python
earthquake_df = gmaps.datasets.load_dataset_as_df('earthquakes')
earthquake_df.head()
```

```python
locations = earthquake_df[['latitude', 'longitude']]
weights = earthquake_df['magnitude']
fig = gmaps.figure()
fig.add_layer(gmaps.heatmap_layer(locations, weights=weights))
fig
```

```python
myGMap = gmaps.figure(zoom_level=4, center=[37.9943708, -1.1275828])

location = []
locationInfo = []
weights=[]
for index, point  in blackshapes_points.iterrows():
    #location.append({'name': point.Address, 'location': [point.lat, point.long]})
    location.append([point.lat, point.long])
    locationInfo.append(point.Address)
    weights.append(point.numAccident)
    #gmaps.Marker(location = [point.lat, point.long],popup=point.Address).add_to(marker_clusterGMaps)
markers = gmaps.marker_layer(locations=location, info_box_content=locationInfo)
gmaps.MarkerOptions()
myGMap.add_layer(markers)
myGMap
```

```python
fig = gmaps.figure(zoom_level=4, center=[37.9943708, -1.1275828])
fig.add_layer(gmaps.heatmap_layer(location, weights=weights))
fig
```

```python
trafficData = pd.read_csv("traffic.csv")
trafficData.head(10)
```

```python
trafficPoss = trafficData[['sensorId','coordinates.lat', 'coordinates.long','speed','observationTime']]
trafficPoss.columns = ['sensorId','lat', 'long', 'speed', 'date']
trafficPoss.loc[(trafficPoss['speed'] > 0)][['sensorId']].head(10000)
```

```python
trafficPoss.loc[(trafficPoss['sensorId'] == 5015966)].head(10)
```

```python
myMap = folium.Map(location=(37.9943708, -1.1275828), zoom_start=6)
marker_cluster = MarkerCluster().add_to(myMap)

points = []
for index, point  in trafficPoss.loc[(trafficPoss['sensorId'] == 5015931)].iterrows():
    folium.Marker(location = [point.lat, point.long],popup='vehicle : ' + str(point.sensorId) + ' \n FECHA: '+str(point.date)).add_to(marker_cluster)
    points.append([point.lat, point.long])
folium.PolyLine(points, color="red", weight=2.5, opacity=1).add_to(myMap)
display(myMap)
```

```python
myGMap = gmaps.figure()

location = []
locationInfo = []
for index, point  in  trafficPoss.loc[(trafficPoss['sensorId'] == 5015966)].iterrows():
    location.append([point.lat, point.long])
    locationInfo.append('id: ' + str(point.sensorId) + '</br> FECHA: '+ point.date)
markers = gmaps.marker_layer(locations=location, info_box_content=locationInfo)
gmaps.MarkerOptions()
myGMap.add_layer(markers)
myGMap
```
